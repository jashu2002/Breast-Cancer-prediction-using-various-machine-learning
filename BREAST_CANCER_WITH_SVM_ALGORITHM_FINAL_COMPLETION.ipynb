```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
#from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
from sklearn.datasets import load_breast_cancer

# Load the breast cancer dataset
data = load_breast_cancer()
#data=pd.read_csv("/content/data.csv")
print(data)
```

```python

# Convert the dataset into a pandas DataFrame using the data and feature names
df = pd.DataFrame(data.data, columns=data.feature_names)

# Add the target variable to the DataFrame
df['diagnosis'] = data.target

# Convert the 'diagnosis' column to 'B' and 'M' (Optional if you need this step for some reason)
df['diagnosis'] = df['diagnosis'].map({0: 'B', 1: 'M'})

# Now, convert the 'diagnosis' column to integers (This is unnecessary if you skip the previous step)
df['diagnosis'] = df['diagnosis'].map({'B': 0, 'M': 1}).astype(int)

'''for column in df.columns[:-1]:  # Exclude the 'diagnosis' column
    df[column] = df[column].astype('float32')'''

print(df.dtypes)

# Display the first few rows of the DataFrame
print(df.head())
```
